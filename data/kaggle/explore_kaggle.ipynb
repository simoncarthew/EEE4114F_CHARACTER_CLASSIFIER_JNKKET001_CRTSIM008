```python
import pandas as pd

# Load the dataset
data = pd.read_csv('A_Z Handwritten Data.csv')

# Define the number of samples to keep for each class
samples_per_class = 1000

# List to store downsampled data
downsampled_data = []

# Loop through each class
for i in range(26):
    # Select rows corresponding to the current class
    class_data = data[data.iloc[:, 0] == i]
    
    # Randomly select samples_per_class rows from the current class
    sampled_class_data = class_data.sample(n=samples_per_class, random_state=42)
    
    # Append the sampled data to downsampled_data list
    downsampled_data.append(sampled_class_data)

# Concatenate downsampled data for all classes
downsampled_data = pd.concat(downsampled_data)

# Save the downsampled dataset to another CSV
downsampled_data.to_csv('baby_kaggle.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from PIL import Image

# Load the dataset
data = pd.read_csv('A_Z Handwritten Data.csv')

# Get unique classes
unique_classes = data.iloc[:, 0].unique()

# List to store one record from each class
one_record_per_class = []

# Select one record from each class
for cls in unique_classes:
    class_data = data[data.iloc[:, 0] == cls]
    one_record_per_class.append(class_data.iloc[0])  # Select the first record from each class

# Save one record from each class as images
for idx, record in enumerate(one_record_per_class):
    # Extract pixel columns (28x28 image)
    pixel_columns = record.index[1:]  # Exclude the first column (classification)
    
    # Extract pixel values and reshape to 28x28
    image_data = record[pixel_columns].values.astype(np.uint8).reshape(28, 28)
    
    # Create image from pixel values
    image = Image.fromarray(image_data)
    
    # Save the image as a JPG file
    image.save(f"image_{idx}.jpg")

print("Images saved successfully.")
```

```python
# load MNIST training and test data
def getKaggleData(file_path, test_size=0.2, test = True):
    # Read the data from the file path
    data = pd.read_csv(file_path)

    # Extract the image data into X and labels into y
    y = data['0']
    X = data.drop('0', axis=1)
    X = X.astype(float)

    # Split into training and testing data
    X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=test_size, random_state=42)

    # Create validation data
    if test:
        X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=test_size, random_state=42)
    else:
        X_valid, y_valid = None, None

    # Reshape the image data into 2D arrays
    X_train = X_train.reshape(-1, 28 * 28)
    X_valid = X_valid.reshape(-1, 28 * 28) if X_valid is not None else None
    X_test = X_test.reshape(-1, 28 * 28)

    return X_train, y_train, X_valid, y_valid, X_test, y_test
```

```python
X_train, y_train, X_valid, y_valid, X_test, y_test = getKaggleData('baby_kaggle.csv', test_size = 0.2, test = True)
```

```python
import pandas as pd
import numpy as np
from PIL import Image
import os

# Load the dataset
data = pd.read_csv('/home/simon/OneDrive/University/Fourth_Year/EEE4114F/Project/EEE4114F_CHARACTER_CLASSIFIER_JNKKET001_CRTSIM008/data/hand_crafted/test_data.csv')

# Get unique classes
unique_classes = data.iloc[:, 0].unique()

# list to store 20 records per class
all_class_data = []

# Select 10 records from each class
for cls in unique_classes:
    class_records = []
    class_data = data[data.iloc[:, 0] == cls]
    for i in range(10):
        class_records.append(class_data.iloc[i])  # Select the first record from each class
    all_class_data.append(class_records)

# define all class names
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

for class_num, class_records in enumerate(all_class_data):
    # Create class directory path
    directory_path = os.path.join(os.getcwd(), alphabet[class_num])
    
    # Check if directory already exists
    if not os.path.exists(directory_path):
        # If directory does not exist, create it
        os.makedirs(directory_path)
        print(f"Created directory: {directory_path}")
    else:
        print(f"Directory already exists: {directory_path}")

    for idx, record in enumerate(class_records):
        # Extract pixel columns (28x28 image)
        pixel_columns = record.index[1:]  # Exclude the first column (classification)
        
        # Extract pixel values and reshape to 28x28
        image_data = record[pixel_columns].values.astype(np.uint8).reshape(28, 28)
        
        # Create image from pixel values
        image = Image.fromarray(image_data)
        
        # Save the image as a JPG file
        image.save(directory_path + f"/image_{idx}.jpg")

print("Images saved successfully.")
```

```python
import os
from PIL import Image
import numpy as np

# convert directory of images into csv
def JPGsToCSV(dir_path,csv_path):
    # define the alphabet
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

    # initialize columns
    columns = []
    for i in range(785):
        columns.append(i)
    
    # initialize empty lits of lists for the data
    data = []

    # loop through every letter/class in alphabet
    for class_num, letter in enumerate(alphabet):
        # set the class directory path
        class_dir_pth = dir_path + "/" + letter

        # get all the jpg files in the class directory
        files = os.listdir(class_dir_pth)
        jpg_files = [file for file in files if file.lower().endswith('.jpg')]

        for file in jpg_files:
            # Open the JPEG image
            image = Image.open(class_dir_pth + "/" + file)
            
            # Convert the image to grayscale
            image = image.convert("L")
            
            # Convert the image to a numpy array
            image_array = np.array(image)
            
            # Flatten the pixel values
            image = image_array.flatten().tolist()

            # add class to the front of the image
            image.insert(0, class_num)

            data.append(image)
    
    df = pd.DataFrame(data,columns=columns)
    df.to_csv(csv_path,index=False)

JPGsToCSV('/home/simon/OneDrive/University/Fourth_Year/EEE4114F/Project/EEE4114F_CHARACTER_CLASSIFIER_JNKKET001_CRTSIM008/data/hand_crafted','/home/simon/OneDrive/University/Fourth_Year/EEE4114F/Project/EEE4114F_CHARACTER_CLASSIFIER_JNKKET001_CRTSIM008/data/hand_crafted/test_data.csv')
```
